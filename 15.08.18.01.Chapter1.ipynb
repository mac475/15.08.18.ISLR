#1. Introduction
##An Overview of Statistical Learning
* Statistical learning은 data를 이해하기 위한 여러가지 tools의 집합
    - supervised statistical learning : input에 대한 output기반의 predicting/ estimating을 위한 model building
    - unsupervised statistical learning : input은 존재하나 supervising을 위한 output이 없는 경우 (하지만, data의 관계/ 구조파악 가능)

###Wage Data, Stock Market Data
* regression problem
    - continuos, quantitative한 output을 predicting
* classification problem
    - non-numerical, categorical, qualitative한 output을 predicting

####Gene Expression Data
* clustering problem
    - input만 존재하고 output이 없는 data를 각 observation이 가지고 있는 특징에 근거하여 grouping

##A Brief History of Statistical Learning
* Gauss의 논문 : method of least squares → 현재의 linear regression으로 발전
* Linear regression : quantitative value를 predicting
* Linear discriminant analysis, logistic regression : qualitative value를 predicting

##Notation
* 전치행렬 : transposed matrix
    $$
        X^T = 
                \left(
                    \begin{matrix}
                        x_{11} & x_{21} & \cdots & x_{n1} \\
                        x_{12} & x_{22} & \cdots & x_{n2} \\
                        \vdots \\
                        x_{1p} & x_{2p} & \cdots & x_{np}
                    \end{matrix}
                \right)
        \\
        \\
        x_i^T = (x_{i1} \quad x_{i2} \quad \cdots x_{ip})
    $$
